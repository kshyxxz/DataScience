```python
import csv
import random

# Expanded pools for high variation
first_names = ["Aarav", "Elena", "Kwame", "Sora", "Isabella", "Dmitri", "Fatima", "Lachlan", "Mei", "Santiago", "Amara", "Finn", "Zahra", "Mateo", "Priya", "Yuki", "Arjun", "Sloan", "Hassan", "Ingrid", "Xavier", "Thao", "Malik", "Freja", "Diego"]
last_names = ["Patel", "Rodriguez", "Osei", "Tanaka", "Rossi", "Volkov", "Al-Sayed", "McKenzie", "Lin", "Morales", "Okafor", "O'Connor", "Nemati", "Jensen", "Sharma", "Watanabe", "Singh", "Vance", "Abadi", "Larsson", "Dupont", "Nguyen", "Belhadj", "MÃ¼ller", "Silva"]
domains = ["@gmail.com", "@outlook.com", "@university.edu", "@protonmail.com", "@icloud.com"]
streets = ["Evergreen Terrace", "Market St", "High Street", "Cherry Blossom Ln", "Piazza Way", "Birch Ave", "Desert Rose Dr", "Highland Rd", "Bamboo Grove", "Sunset Blvd"]
cities = ["Springfield", "Philadelphia", "Boston", "Seattle", "Anchorage", "Phoenix", "San Francisco", "Los Angeles", "Chicago", "Austin"]

with open('students_varied.csv', 'w', newline='', encoding='utf-8') as f:
    writer = csv.writer(f)
    writer.writerow(["student_id", "name", "email", "phone", "dob", "address"])
    
    for i in range(1, 1001):
        fn = random.choice(first_names)
        ln = random.choice(last_names)
        full_name = f"{fn} {ln}"
        email = f"{fn.lower()}.{ln.lower()}{random.randint(1,99)}{random.choice(domains)}"
        phone = f"{random.randint(100, 999)}-{random.randint(100, 999)}-{random.randint(1000, 9999)}"
        dob = f"{random.randint(2003, 2007)}-{random.randint(1, 12):02d}-{random.randint(1, 28):02d}"
        address = f"{random.randint(1, 9999)} {random.choice(streets)}, {random.choice(cities)}"
        
        writer.writerow([f"STU-{i}", full_name, email, phone, dob, address])

print("Generated students_varied.csv with 1000 records!")
```
